```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("summer.csv")
df.isnull().sum()
```

## 1. In how many cities is Olympics held?


```python
cities = len(df['City'].unique())
print("The number of cities that have held Olympics are: ",cities)
```

## 2. Which sport is having most number of Gold Medals so far? (Top 5)


```python
df_1 = df[df['Medal'] == 'Gold']                                                 # Creating a new dataframe with only gold medals
data = []

for i in df_1['Sport'].unique():                                             
  data.append([i, len(df_1[df_1['Sport'] == i])])                                # Appending the sport and the number of 
                                                                                 # gold medals to the list

data = pd.DataFrame(data, columns = ['Sport','No.of gold medals'])               # Converting it back to dataframe
data = data.sort_values(by = 'No.of gold medals', ascending = False)
data = data.head()
print(data)
print()
print("The sport having most number of gold medals is:",data.iat[0,0])
print()
data.plot(x = 'Sport', y = 'No.of gold medals', kind = 'bar', color = 'y',       # Plotting the graph
          title = 'Sports having most number of Gold Medals', figsize = (7,7))
```

## 3. Which sport is having most number of medals so far? (Top 5)

```python
data_3 = []

for i in df['Sport'].unique():                                                   # Appending the name of the sport
  data_3.append([i, len(df[df['Sport'] == i])])                                  # and the number of medals in each sport

data_3 = pd.DataFrame(data_3, columns = ['Sport','No.of medals'])                # Converting back to dataframe
data_3 = data_3.sort_values(by = 'No.of medals', ascending = False)              # Sorting according to the Number of medals 
data_3 = data_3.head()
print(data_3)
print()
print("The sport having most number of medals is:",data_3.iat[0,0])
print()
data_3.plot(x = 'Sport', y = 'No.of medals', kind = 'bar', 
            title = 'Sports having the most number of medals', 
            figsize = (10,8))                                                    # Plotting Graph 
  
```

## 4. Which player has won most number of medals? (Top 5)

```python
data_4 = []

for i in df['Athlete'].unique():                                                 # Appending the Athelete and number of medals to a list
  data_4.append([i, len(df[df['Athlete'] == i])])

data_4 = pd.DataFrame(data_4, columns = ['Athlete','No. of medals'])
data_4 = data_4.sort_values(by = 'No. of medals', ascending = False)
data_4 = data_4.head()                                                           # saving the top 5 rows of the dataframe 
print(data_4)
print()
print("The player having most number of medals is:",data_4.iat[0,0])
print()
data_4.plot(x = 'Athlete', y = 'No. of medals', kind = 'bar', color = 'g',       # Plotting the graph
            title = 'Players that have won the most number of medals', 
            figsize = (10,8))
 
```

## 6. In which year India won first Gold Medal in Summer Olympics?

```python
df_3 = df_2[df_2['Country'] == 'IND'].sort_values(by = 'Year', ascending = True) # Searching for 'IND' in the dataframe and
answer = df_3.iat[0,0]                                                           # arranging them in ascending order and printing 
print("India won the first gold medal in the year: ",answer)                     # the first year
```

## 7. Which event is most popular in terms on number of players? (Top 5)

```python
data_7 = []

for i in df['Event'].unique():                                                   # Appending Event and number of player to the list
  data_7.append([i, len(df[df['Event'] == i])])

data_7 = pd.DataFrame(data_7, columns = ['Event','No. of players'])
data_7 = data_7.sort_values(by = 'No. of players', ascending = False)            # sorting values
data_7 = data_7.head()
print(data_7)
print()
print("The event which is most popular in terms of number of players is: ",
      data_7.iat[0,0])
print()
data_7.plot(x = 'Event', y = 'No. of players', kind = 'bar', color = 'orange',   # Plotting graph
            title = 'Events that have had the most number of players', 
            figsize = (10,8))
```
